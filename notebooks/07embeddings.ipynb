```python
A, B, C = "dog", "cat", "horse"
D = "zebra"
```

```python
vec = lambda word: model[word]
vec_A, vec_B, vec_C = vec(A), vec(B), vec(C)
vec_D = vec(D)
```

```python
diff = vec_B - vec_A
```

```python
# Base del plano
e1 = diff / np.linalg.norm(diff)  # eje X
```

```python
predicted_word, score = model.most_similar(positive=[C, B], negative=[A], topn=1)[0]
vec_Pred = vec(predicted_word)
print(predicted_word)
```

```python
# Construir eje ortogonal (eje Y del plano)
v_tmp = vec_C - vec_A
proj_on_e1 = np.dot(v_tmp, e1) * e1
e2 = v_tmp - proj_on_e1
e2 /= np.linalg.norm(e2)
```

```python
# Función de proyección al plano (e1, e2)
def project(v):
    return np.array([np.dot(v, e1), np.dot(v, e2)])
```

```python
# Proyección de los vectores al plano
points = {
    A: vec_A,
    B: vec_B,
    C: vec_C,
    D: vec_D,
    predicted_word: vec_Pred  # puede coincidir con "queen"
}

# Graficar
plt.figure(figsize=(7, 7))
for word, vector in points.items():
    x, y = project(vector)
    plt.scatter(x, y)
    plt.text(x + 0.01, y + 0.01, word)

plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid()
plt.title(f"Analogía: {A} → {B} :: {C} → {predicted_word} (score={score:.2f})")
plt.axis('equal')
plt.show()
```
